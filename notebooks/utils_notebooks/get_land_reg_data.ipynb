```python
import pandas as pd
from pathlib import Path
from config.config import DATA_RAW
```

```python
# Get headers from txt file
land_reg_root = DATA_RAW / "land_registry"
land_cols_path = Path(land_reg_root) / "land_reg_cols.txt"

with open(land_cols_path, "r") as f:
    headers = [line.strip() for line in f.readlines()]

headers
```

```python
# Read in csv files and add header to each of them
land_reg_2022 = Path(land_reg_root) / "pp-2022.csv"
land_reg_2023 = Path(land_reg_root) / "pp-2023.csv"
land_reg_2024 = Path(land_reg_root) / "pp-2024.csv"

df = pd.read_csv(filepath_or_buffer=land_reg_2022, header=None)
df.columns = headers
df.to_csv(land_reg_2022 , index=False)
```
